```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('/content/drive/MyDrive/students.csv')
```

```python
dfcut = df[['Growth', 'Weight', 'Sex']]
```

```python
dfcut = dfcut.dropna()
```

```python
sns.scatterplot(data=dfcut, x='Weight', y='Growth', hue='Sex')
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(dfcut[['Growth', 'Weight']].values.reshape(-1, 2))
arr = scaler.transform(dfcut[['Growth', 'Weight']].values.reshape(-1, 2))
```

```python
model = KNeighborsClassifier(n_neighbors=1)
model.fit(arr, y=dfcut['Sex'].values)
```

```python
test = pd.read_csv('/content/students_test.csv')
```

```python
cuttest = test[['Growth', 'Weight', 'Sex']]
```

```python
cuttest = cuttest.dropna()
```

```python
arrtest = scaler.transform(cuttest[['Growth', 'Weight']].values.reshape(-1, 2))
```

```python
cuttest['Predicted'] = model.predict(arrtest)
```

```python
dfdf = pd.read_csv('/content/drive/MyDrive/students.csv')[['Hair length','Shoe size', 'Sex']]
```

```python
dfdf = dfdf.dropna()
```

```python
sns.scatterplot(data=dfdf, x='Hair length', y='Shoe size', hue='Sex')
```

```python
scaler.fit(dfdf[['Hair length','Shoe size']].values.reshape(-1, 2))
arr1 = scaler.transform(dfdf[['Hair length','Shoe size']].values.reshape(-1, 2))
```

```python
model1 = KNeighborsClassifier(n_neighbors=3)
model1.fit(arr1, y=dfdf['Sex'].values)
```

```python
dftest = pd.read_csv('/content/students_test.csv')[['Hair length','Shoe size', 'Sex']]
```

```python
dftest = dftest.dropna()
```

```python
arr2 = scaler.transform(dftest[['Hair length','Shoe size']].values.reshape(-1, 2))
```

```python
dftest['Predicted'] = model1.predict(arr2)
```

```python
pd.crosstab(dftest['Predicted'], dftest['Sex'])
```
